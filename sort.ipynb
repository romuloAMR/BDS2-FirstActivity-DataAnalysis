# Data Analysis

#### Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Importing Data frame

```python
data = pd.read_csv("Sort.csv")
data.info()
data.head()
```

#### Partitioning the data

```python
bubble = ["BubbleSortI", "BubbleSortR"]
merge = ["MergeSortI", "MergeSortR"]
quick = ["QuickSortI", "QuickSortR"]
algorithms = bubble + merge + quick
sizes = [1000, 10000, 100000]
means = data.groupby(['Algorithm', 'Size'])['Time'].mean().unstack()
means
```

#### Bar graphs

##### BubbleSort

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4), sharey=True)

for ax, algorithm in zip(axes, bubble):
    y_values = means.loc[algorithm]
    y_values.plot(kind='bar', ax=ax, width=0.5)

    ax.set_ylabel('Means (Time in seconds)')
    ax.set_xlabel('Sizes')
    ax.set_title(algorithm)
    ax.set_xticks(range(len(sizes)))
    ax.set_xticklabels(sizes)

plt.tight_layout()
plt.show()
```

##### MergeSort

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4), sharey=True)

for ax, algorithm in zip(axes, merge):
    y_values = means.loc[algorithm]
    y_values.plot(kind='bar', ax=ax, width=0.5)

    ax.set_ylabel('Means (Time in seconds)')
    ax.set_xlabel('Sizes')
    ax.set_title(algorithm)
    ax.set_xticks(range(len(sizes)))
    ax.set_xticklabels(sizes)

plt.tight_layout()
plt.show()
```

##### QuickSort

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4), sharey=True)

for ax, algorithm in zip(axes, quick):
    y_values = means.loc[algorithm]
    y_values.plot(kind='bar', ax=ax, width=0.5)

    ax.set_ylabel('Means (Time in seconds)')
    ax.set_xlabel('Sizes')
    ax.set_title(algorithm)
    ax.set_xticks(range(len(sizes)))
    ax.set_xticklabels(sizes)

plt.tight_layout()
plt.show()
```

##### Size comparison

```python
for size in sizes:
    fig, ax = plt.subplots(figsize=(10, 5))
    y_values = means[size]
    y_values.plot(kind='bar', ax=ax, width=0.5)

    ax.set_ylabel('Means (Time in seconds)')
    ax.set_xlabel('Algorithms')
    ax.set_title(f'Size: {size}')
    ax.set_xticks(range(len(algorithms)))
    ax.set_xticklabels(algorithms)

    plt.tight_layout()
    plt.show()
```
